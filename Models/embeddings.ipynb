```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import umap.umap_ as umap
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# --- Carica dati ---
# Assumi file CSV "news.csv" con colonne: title, excerpt, date
df = pd.read_csv("../Data/raw/df_300.csv")

# Pulizia minima: rimuovi righe senza titolo o excerpt
df = df.dropna(subset=['title', 'excerpt', 'date'])

# --- Prepara testi ---
texts = (df['title'] + " " + df['excerpt']).tolist()

# --- Calcola embedding ---
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(texts, show_progress_bar=True)

# --- Riduzione dimensionale con UMAP ---
reducer = umap.UMAP(n_components=2, random_state=42)
embedding_2d = reducer.fit_transform(embeddings)

# --- Prepara colori per date ---
unique_dates = df['date'].unique()
colors = cm.get_cmap('tab20', len(unique_dates))
date_to_color = {date: colors(i) for i, date in enumerate(unique_dates)}
point_colors = df['date'].map(date_to_color)

# --- Plot ---
plt.figure(figsize=(12,8))
plt.scatter(embedding_2d[:,0], embedding_2d[:,1], c=point_colors, alpha=0.7, s=50)

# Legenda date
for date, color in date_to_color.items():
    plt.scatter([], [], color=color, label=date)
plt.legend(title="Date", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title("News embeddings visualized with UMAP")
plt.xlabel("UMAP 1")
plt.ylabel("UMAP 2")
plt.tight_layout()
plt.show()

# --- Salva risultato ---
df_out = df.copy()
df_out['embedding_2d_1'] = embedding_2d[:,0]
df_out['embedding_2d_2'] = embedding_2d[:,1]
df_out.to_csv("news_embeddings_umap.csv", index=False)

print("Embedding calcolati e salvati in news_embeddings_umap.csv")
```
